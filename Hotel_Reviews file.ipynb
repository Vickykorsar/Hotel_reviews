```python
x=x.drop(columns='index',axis=1)
x
```

```python
X=[]
for i in range(len(x)):
    a=x['reviews.text'].loc[i]
    for j in sent_tokenize(a):
        b=re.sub('[^A-Za-z0-9]',' ',j)
        b=b.lower()
        b=b.split()
        b=[P.stem(k) for k in b if k not in stopwords.words('english')]
        b=' '.join(b)
        X.append(b)
        
```

```python
X
```

```python
T=TfidfVectorizer(ngram_range=(2,6),max_features=40)
C=CountVectorizer(max_features=500)
TR=T.fit_transform(X).toarray()
```

```python
T
```

```python
C
```

```python
TR1=pd.DataFrame(TR,columns=T.get_feature_names_out())
TR1
```

```python
TR1['reviews.rating']=x['reviews.rating']
```

```python
TR1
```

```python
mode_value = TR1['reviews.rating'].mode() 
TR1['reviews.rating'].fillna(mode_value, inplace=True)
```

```python
TR1["reviews.rating"].replace({4.078:1},inplace=True)
```

```python
F=TR1.iloc[:,:-1]
T=TR1.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(F,T)
```

```python
from sklearn.svm import SVC
S=SVC()
model=S.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(classification_report(pred,y_test))
```
